```python
import pandas as pd

#Adult Literacy Rate

df_adult = pd.read_csv("https://ourworldindata.org/grapher/literacy-rate-adults.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})

#Rename columns
df_adult.rename(columns={'entity':'country','adult_literacy_rate__population_15plus_years__both_sexes__pct__lr_ag15t99':'adult_literacy'},inplace=True)

# Filter year
df_adult = df_adult[df_adult['year'] >= 1990]

# Reset index
df_adult.reset_index(drop=True, inplace=True)


# Sort
df_adult.sort_values(by='country',inplace=True)

# View
df_adult.head()





```

```python
#Youth Literacy Rate
df_youth=pd.read_csv("https://ourworldindata.org/grapher/literacy-rate-of-young-men-and-women.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})

# Rename columns
df_youth.rename(columns={'entity':'country','youth_literacy_rate__population_15_24_years__male__pct__lr_ag15t24_m':'youth_male','youth_literacy_rate__population_15_24_years__female__pct__lr_ag15t24_f':'youth_female'
}, inplace=True)



# Filter year
df_youth = df_youth[df_youth['year'] >= 1990]
# Reset index
df_youth.reset_index(drop=True, inplace=True)
# Sort
df_youth.sort_values(by='country', inplace=True)

# View
df_youth.head()

```

```python

#Illiterate Population
df_illiterate=pd.read_csv("https://ourworldindata.org/grapher/literate-and-illiterate-world-population.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})


# Rename columns
df_illiterate.rename(columns={'entity':'country'}, inplace=True)

# Filter year
df_illiterate = df_illiterate[df_illiterate['year'] >= 1990]
# Reset index
df_illiterate.reset_index(drop=True, inplace=True)
# Sort
df_illiterate.sort_values(by='country', inplace=True)


# View
df_youth.head()
```

```python
#GDP per Capita

df_gdp=pd.read_csv("https://ourworldindata.org/grapher/gdp-per-capita-worldbank.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})




# Rename columns
df_gdp.rename(columns={'entity':'country','ny_gdp_pcap_pp_kd': 'gdp_per_capita'
}, inplace=True)

# Filter year
df_gdp = df_gdp[df_gdp['year'] >= 1990]
# Reset index
df_gdp.reset_index(drop=True, inplace=True)
# Sort

df_gdp = df_gdp.sort_values(by=['country','year'])



# View
df_gdp.head()
```

```python
#Average Years of Schooling

df_school=pd.read_csv("https://ourworldindata.org/grapher/literacy-rates-vs-average-years-of-schooling.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})

# Rename columns
df_school.rename(columns={'entity':'country','mf_youth_and_adults__15_64_years__average_years_of_education':'avg_schooling'
}, inplace=True)

# Filter year
df_school = df_school[df_school['year'] >= 1990]
# Reset index
df_school.reset_index(drop=True, inplace=True)
# Sort

df_school.sort_values(by='country',inplace=True)
#df_gdp = df_gdp.sort_values(by=['country','year'])

# View
df_school.head()


```

```python
#Merge

df_literacy = pd.merge(
    df_adult,
    df_youth,
    on=['country','code','year'],
    how='inner'
)

df_literacy.head()
```

```python
#separate
df_illiterate = df_illiterate.copy()
df_illiterate.head()
```

```python
#Merge
df_gdp_schooling = pd.merge(
    df_gdp,
    df_school,
    on=['country','code','year'],
    how='inner'
)
df_gdp_schooling.head()
```

```python
#Data cleaning for literacy
# Check shape
df_literacy.shape

# Fill missing 'code'
if 'code' in df_literacy.columns:
    df_literacy['code'] = df_literacy['code'].fillna("UNK")

# Fix owid_region (IMPORTANT)
if 'owid_region' in df_literacy.columns:
    df_literacy['owid_region'] = df_literacy['owid_region'].fillna("Unknown")

# Final check
df_literacy.info()
```

```python
#Data cleaning for illiterate
df_illiterate.shape
df_illiterate.columns
# Safe drop
df_illiterate.drop(columns=['code','literacy_rate'], errors='ignore', inplace=True)

# Filter year
df_illiterate = df_illiterate[df_illiterate['year'] >= 1990]

# Remove duplicates
df_illiterate.drop_duplicates(inplace=True)

# Sort
df_illiterate = df_illiterate.sort_values(by=['country','year'])

# Reset index
df_illiterate.reset_index(drop=True, inplace=True)

# Check
df_illiterate.info()
df_illiterate.head()
```

```python
#Data cleaning for gdp_schooling
df_gdp_schooling.shape
df_gdp_schooling.columns

# Drop
df_gdp_schooling.drop(columns=[
    'code',
    'owid_region_x',
    'owid_region_y',
    'population_historical',
    'literacy_rate'
], errors='ignore', inplace=True)
#Remove missing values
df_gdp_schooling.dropna(inplace=True)
# Filter year
df_gdp_schooling = df_gdp_schooling[df_gdp_schooling['year'] >= 1990]
# Remove duplicates
df_gdp_schooling.drop_duplicates(inplace=True)
# Sort
df_gdp_schooling = df_gdp_schooling.sort_values(by=['country','year'])
#Reset index
df_gdp_schooling.reset_index(drop=True, inplace=True)

df_gdp_schooling.info()

df_gdp_schooling.head()
```

```python
#Feature engineering
#Literacy Features
# 1. Gender Gap (male - female)
df_literacy['gender_gap'] = df_literacy['youth_male'] - df_literacy['youth_female']

# 2. Youth Average
df_literacy['youth_avg'] = (df_literacy['youth_male'] + df_literacy['youth_female']) / 2

# 3. Literacy Growth (year-wise increase)
df_literacy = df_literacy.sort_values(by=['country','year'])

df_literacy['literacy_growth'] = df_literacy.groupby('country')['adult_literacy'].pct_change() * 100
df_literacy.fillna({'literacy_growth': 0}, inplace=True)
#df_literacy['literacy_growth'].fillna(0, inplace=True)

# Check
df_literacy.head()
```

```python
# Rename for clarity
df_illiterate.rename(columns={'illiteracy_rate': 'illiteracy_percent'}, inplace=True)

# Check
df_illiterate.head()
```

```python
# GDP per schooling year
df_gdp_schooling['gdp_per_schooling'] = df_gdp_schooling['gdp_per_capita'] / df_gdp_schooling['avg_schooling']

# Check
df_gdp_schooling.head()
```

```python
# Merge literacy + GDP data
df_edu = pd.merge(df_literacy, df_gdp_schooling, on=['country','year'], how='inner')

# Check
df_edu.head()
```

```python
# Normalize values
df_edu['norm_literacy'] = df_edu['adult_literacy'] / 100
df_edu['norm_schooling'] = df_edu['avg_schooling'] / df_edu['avg_schooling'].max()

# Create index
df_edu['education_index'] = (df_edu['norm_literacy'] + df_edu['norm_schooling']) / 2

# Final check
df_edu.head()
```

```python
#1)Top 5 countries with highest adult literacy in 2020

q1 = """
SELECT country, adult_literacy
FROM literacy_rates
WHERE year = 2020
ORDER BY adult_literacy DESC
LIMIT 5;
"""
import pandas as pd
pd.read_sql(q1, conn)
df_q1 = pd.read_sql(q1, conn)
df_q1
```

```python
#2)Female youth literacy < 80%
q2="""SELECT country, youth_female
FROM literacy_rates
WHERE youth_female < 80;"""
pd.read_sql(q2, conn)
df_q2 = pd.read_sql(q2, conn)
df_q2


```

```python
#3)average
q3="""SELECT country, AVG(adult_literacy) AS avg_literacy
FROM literacy_rates
GROUP BY country;"""
pd.read_sql(q3, conn)
df_q3 = pd.read_sql(q3, conn)
df_q3

```

```python
#4)illiterate

q4="""SELECT country, illiteracy_percent
FROM illiterate_population
WHERE year = 2000 AND illiteracy_percent > 20;"""
pd.read_sql(q4, conn)
df_q4 = pd.read_sql(q4, conn)
df_q4

```

```python
#5)trend illiterate

q5="""SELECT year, illiteracy_percent
FROM illiterate_population
WHERE country = 'India' AND year BETWEEN 2000 AND 2020
ORDER BY year;"""
pd.read_sql(q5, conn)
df_q5 = pd.read_sql(q5, conn)
df_q5


```

```python
#6)top 10 countries with highest illiteracy (latest year)
q6="""SELECT country, illiteracy_percent
FROM illiterate_population
WHERE year = (SELECT MAX(year) FROM illiterate_population)
ORDER BY illiteracy_percent DESC
LIMIT 10;"""

pd.read_sql(q6, conn)
df_q6 = pd.read_sql(q6, conn)
df_q6


```

```python
#7)Schooling > 7 and GDP < 5000
q7="""SELECT country, avg_schooling, gdp_per_capita
FROM gdp_schooling
WHERE avg_schooling > 7 AND gdp_per_capita < 5000;"""
pd.read_sql(q7, conn)
df_q7 = pd.read_sql(q7, conn)
df_q7


```

```python
#8)countries by GDP per schooling (2020)
q8="""SELECT country, gdp_per_schooling,
RANK() OVER (ORDER BY gdp_per_schooling DESC) AS rank_no
FROM gdp_schooling
WHERE year = 2020;"""

pd.read_sql(q8, conn)
df_q8 = pd.read_sql(q8, conn)
df_q8


```

```python
#9)Global average schooling per year
q9="""SELECT year, AVG(avg_schooling) AS avg_schooling
FROM gdp_schooling
GROUP BY year
ORDER BY year;"""

pd.read_sql(q9, conn)
df_q9 = pd.read_sql(q9, conn)
df_q9


```

```python
#10)High GDP but low schooling (<6)
q10="""SELECT g.country, g.gdp_per_capita, g.avg_schooling
FROM gdp_schooling g
WHERE g.year = 2020 AND g.avg_schooling < 6
ORDER BY g.gdp_per_capita DESC
LIMIT 10;"""

pd.read_sql(q10, conn)
df_q10 = pd.read_sql(q10, conn)
df_q10


```

```python
#11)High illiteracy but high schooling (>10)



q11 = """
SELECT g.country, g.avg_schooling, i.illiteracy_percent
FROM gdp_schooling g
JOIN illiterate_population i
ON g.country = i.country AND g.year = i.year
WHERE g.avg_schooling > 8;
"""
pd.read_sql(q11, conn)
df_q11= pd.read_sql(q11, conn)
df_q11



```

```python
#12)Literacy vs GDP trend (example: India)

q12 = """
SELECT l.year, l.adult_literacy, g.gdp_per_capita
FROM literacy_rates l
LEFT JOIN gdp_schooling g
ON l.country = g.country AND l.year = g.year
WHERE l.country = 'India'
ORDER BY l.year;
"""

pd.read_sql(q12, conn)
df_q12= pd.read_sql(q12, conn)
df_q12


```

```python
#13)Gender gap where GDP > 30000 (2020)
q13="""SELECT l.country, l.youth_male, l.youth_female,
(l.youth_male - l.youth_female) AS gap
FROM literacy_rates l
JOIN gdp_schooling g
ON l.country = g.country AND l.year = g.year
WHERE g.gdp_per_capita > 30000 AND l.year = 2020;"""
pd.read_sql(q13, conn)
df_q13= pd.read_sql(q13, conn)
df_q13
```

```python
%%writefile app.py

import streamlit as st
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt


# Connect SQLite
conn = sqlite3.connect("education.db")

# Sidebar
st.sidebar.title("Navigation")
st.markdown(    """
<style>
[data-testid="stSidebarContent"] {
    color: white;
    background-color:#63e5ff;
}
</style>
""", unsafe_allow_html=True)

page = st.sidebar.radio("Go to", ["SQL Query", "EDA", "Country Profile"])
st.set_page_config(page_title="Global Literacy & Education Trends", layout="wide")
#SQL query executor


if page == "SQL Query":
    st.title("SQL Query Executor")
    st.markdown("""
    <style>

    .stApp {
        background-color:#a575bd

    }
    h1, h2, h3 {
        font-family: 'Arial';
        color: #2c3e50;
    }
    </style>

""", unsafe_allow_html=True)



    query = st.text_area("Enter SQL Query")

    if st.button("Run Query"):
        try:
            df = pd.read_sql(query, conn)
            st.write(df)

            # Simple chart
            if len(df.columns) >= 2:
                st.bar_chart(df.set_index(df.columns[0]))

        except Exception as e:
            st.error(e)
#EDA Visualization
elif page == "EDA":
    st.title("EDA Visualizations")
    st.markdown("""
        <style>
        .stApp {
            background-color: #F38BC8;
        }
        </style>
    """, unsafe_allow_html=True)


    df = pd.read_sql("SELECT * FROM literacy_rates", conn)
    df2 = pd.read_sql("SELECT * FROM gdp_schooling", conn)
    # 1️⃣ Histogram (Distribution)
    st.subheader("Distribution of Adult Literacy")
    fig1, ax1 = plt.subplots()
    ax1.hist(df['adult_literacy'].dropna())
    st.pyplot(fig1)

    # 2️⃣ Bar Chart (Top Countries)
    st.subheader("Youth Literacy Average")
    top = df.sort_values(by='adult_literacy', ascending=False).head(10)

    fig2, ax2 = plt.subplots()
    ax2.bar(top['country'], top['adult_literacy'])
    plt.xticks(rotation=45)
    st.pyplot(fig2)


    # 3️⃣ Line Chart (Trend)
    st.subheader("India Literacy Trend")
    india = df[df['country'] == 'India']

    fig3, ax3 = plt.subplots()
    ax3.plot(india['year'], india['adult_literacy'])
    st.pyplot(fig3)

    # 4️⃣ Scatter Plot (Relationship)
    st.subheader("GDP vs Schooling")
    fig4, ax4 = plt.subplots()
    ax4.scatter(df2['avg_schooling'], df2['gdp_per_capita'])
    ax4.set_xlabel("Schooling")
    ax4.set_ylabel("GDP")
    st.pyplot(fig4)


    # 5️⃣ Gender Gap Analysis
    st.subheader("Gender Gap Distribution")
    df['gender_gap'] = df['youth_male'] - df['youth_female']

    fig5, ax5 = plt.subplots()
    ax5.hist(df['gender_gap'].dropna())
    st.pyplot(fig5)
#Country Profile page
elif page == "Country Profile":
    st.title("Country Profile")
    st.markdown("""
        <style>
        .stApp {
            background-color: #FFAE42
        }
        </style>
    """, unsafe_allow_html=True)

    df = pd.read_sql("SELECT * FROM literacy_rates", conn)

    country = st.selectbox("Select Country", df['country'].unique())

    df_country = df[df['country'] == country]

    st.write(df_country)

    # Line chart
    fig, ax = plt.subplots()
    ax.plot(df_country['year'], df_country['adult_literacy'])
    ax.set_title(f"{country} Literacy Trend")
    st.pyplot(fig)

```
